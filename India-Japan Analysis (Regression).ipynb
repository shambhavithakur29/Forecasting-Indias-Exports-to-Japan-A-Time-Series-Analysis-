REGRESSION ANALYSIS:

Exports_Ind_Jpn = Î²0 + Î²1*(GDP_Jpn) + Î²2*(INR_JPN_ER) + Î²3*(Crude_Oil_Price) + error

Dependent variable: Exports of India to Japan (Exports_Ind_Jpn)
Independent Variables: Gdp of Japan(GDP_Jpn), Exchange rate(INR_JPN_ER), Crude Oil price(Crude_Oil_Price)

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
Z = INDJPN_df[['GDP_Jpn', 'INR_YEN_ER', 'Crude_Oil_Price']]
```

```python
lm = LinearRegression()
```

```python
lm.fit(Z, INDJPN_df['Exports_Ind_Jpn'])
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
Y_hat = lm.predict(Z)
```

```python
plt.figure(figsize=(10, 6))


ax1 = sns.distplot(INDJPN_df['Exports_Ind_Jpn'], hist=False, color="plum", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="skyblue", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Exports')
plt.xlabel('Exports of India to Japan (USD)')
plt.ylabel('Density')

plt.show()
plt.close()
```

```python
# Find the R^2

print('The R-square is: ', lm.score(Z, INDJPN_df['Exports_Ind_Jpn']))
Y_hat = lm.predict(Z)
print('The mean square error of price and predicted value using multifit is: ', \
      mean_squared_error(INDJPN_df['Exports_Ind_Jpn'], Y_hat))
```
